##a. Use forward Euler method to solve the following problem
$$y'=-5y,y(0)=1$$
Solve the problem for $t \in [0,10]$ with $h=0.1,0.4,0.41$

Notice that the forward Euler method have to check it's global stability
$$\lvert 1+hÎ»\lvert <1$$
so we get in this case we must make sure
$$
h<0.4
$$
and the exact solution of this case is $y=e^{-5t}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward_euler(f, y0, t0, T, h):
    N = int(np.ceil((T - t0) / h))
    t = np.linspace(t0, t0 + N*h, N + 1)
    y = np.empty(N + 1, dtype=float)
    y[0] = y0
    for n in range(N):
        y[n+1] = y[n] + h * f(t[n], y[n])
    return t, y
```

#**Here give a stable solution $h=0.1$**

# $h=0.1$

```python
f = lambda t, y: -5.0 * y
y0 = 1.0
t0, T = 0.0, 10.0
h = 0.1
t, y_num = forward_euler(f, y0, t0, T, h)
y_exact = np.exp(-5.0 * t)
abs_err = np.abs(y_num - y_exact)
```

```python
print("  n        t_n          y_euler         y_exact        |error|")
for n in range(min(15, len(t))):
    print(f"{n:3d}  {t[n]:9.4f}   {y_num[n]:12.8f}   {y_exact[n]:12.8f}   {abs_err[n]:10.3e}")
```

```python
plt.figure()
plt.plot(t, y_exact, label="Exact $e^{-5t}$")
plt.plot(t, y_num, "o--", label=f"Euler (h={h})")
plt.xlabel("t"); plt.ylabel("y(t)")
plt.title("Forward Euler for y'=-5y, y(0)=1")
plt.legend(); plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

For $h=0.1$ the solution of forward Euler $u_n=(1+hÎ»)^n=0.5^n$

so the solution must be approach the exact solution.

## b. Use backward Euler method to solve the following problem
$$y'=5y,y(0)=1$$
Solve the problem for $t \in [0,10]$ with $h=0.1,0.4,0.41$

Notice that the backward Euler method we don't need  check it's global stability and the solution
$$
u_n=(\frac{1}{1-hÎ»})^{n}
$$

```python
def backward_euler(lambda_, y0, t0, T, h):
    N_full = int(np.floor((T - t0) / h + 1e-14))
    t = t0 + np.arange(N_full + 1) * h
    y = np.empty(N_full + 1); y[0] = y0

    amp = 1.0 / (1.0 - h * lambda_)
    for n in range(N_full):
        y[n+1] = amp * y[n]

    rem = T - t[-1]
    if rem > 1e-12:
        t = np.append(t, T)
        y = np.append(y, y[-1] / (1.0 - rem*lambda_))
    return t, y
```

#$h=0.1$

```python
h = 0.1
lambda_ = 5.0
y0, t0, T = 1.0, 0.0, 10.0
t, y_num = backward_euler(lambda_, y0, t0, T, h)
y_exact = np.exp(lambda_ * t)
```

```python
print(f"h = {h}")
print(f"y(T) by Backward Euler = {y_num[-1]:.6f}")
print(f"y(T) exact            = {y_exact[-1]:.6f}")
print(f"absolute error        = {abs(y_num[-1]-y_exact[-1]):.3e}")
```

```python
plt.figure()
plt.plot(t, y_exact, label="Exact $e^{5t}$")
plt.plot(t, y_num, "o--", label=f"Backward Euler (h={h})")
plt.xlabel("t"); plt.ylabel("y(t)")
plt.title(f"Backward Euler for y'=5y, h={h}")
plt.legend(); plt.grid(True, alpha=0.3)
plt.show()
```

For $h=0.1$ the solution of forward Euler $u_n=(2)^n$

Each step multiplies by 2, which grows faster than the exact single-step growth $e^{0.5}â1.6487.$
â The numerical solution severely overestimates the true solution (the curve lies far above $e^{5t}$.

For $h=0.1$ the solution of forward Euler $u_n=(\frac{1}{1-2.05})^n$

The solution alternates in sign and decays toward 0, but the exact solution grows rapidly on the order of $e^{50}$.
â An even more qualitatively wrong behavior: it should grow quickly, but instead it oscillates and decays.

For growth problems with $\lambda>0$, the backward Euler method is a poor choice.
When $h\ge 0.2$, we have $1-5h\le 0$, so the perâstep amplification
$$
  u_n=\frac{1}{1-5h}
$$
is negative, causing sign alternation.
Even for $h<0.2$, $\rho(h)$ still \emph{overestimates} the oneâstep growth compared with the exact factor $e^{5h}$; indeed, for small $x>0$,
$$
  \frac{1}{1-x}=1+x+x^{2}+\cdots \;\ge\; e^{x}=1+x+\frac{x^{2}}{2}+\cdots,
$$
so $\rho(h)\ge e^{5h}$ and the numerical solution grows too fast unless $h$ is very small.
Consequently, very small $h$ is required to keep the numerical solution on the correct scale, and moderate $h$ can still distort the qualitative behavior.


#2.Consider solving the following problem using forward Euler method $$y'=y(1-y),y(0)=y_0,0<y_0<1$$ Find the range of $h$ such that the solution is qualitatively correct.

For such question we know that the forward Euler mathod is $u_{n+1}=u_n+hf(t_n,u_n)$ so $$y_{n+1}=y_n+hy_n(1-y_n).$$ If we want the solution qualitatively correct, then we must check the following.

1.   $y_n$ is increasing
2.   $0<y_n<1$
3.   $y_nâ1$

##1. Monotonicity
We can found out that $y_{n+1}-y_n=hy_n(1-y_n)$ so that if we want $y_{n+1}-y_n>0$ this implis that $h>0$.

##2. Upper bound and Lower bound
If $0<h\le 1$ then since  $0<y_n(1-y_n)\le \tfrac14$,
$$y_{n+1}=y_n+h\,y_n(1-y_n)\le y_n+ y_n(1-y_n)=y_n(2-y_n)\le1.$$
If $h>1$ then
$$y_{n+1}=y_n+h\,y_n(1-y_n)=\frac{(1+h)^2}{4h}>1$$
so $0<h\le1$

##3. Convergence
Since $ 0<h\le 1 $ and $\{y_n\}$ has a upper bound 1, so the sequence must converge to some constant $L\in[0,1]$ and $L$ satisfy
$$L=L+hL(1-L)\quad\Rightarrow\quad L\in\{0,1\}.$$
Hence $L=1$

So the range of $h$ is $0<h\le 1$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def forward_euler_logistic(y0, h, T):
    N_full = int(np.floor(T / h + 1e-14))
    t = np.arange(N_full + 1) * h
    y = np.empty(N_full + 1, dtype=float)
    y[0] = y0
    for n in range(N_full):
        y[n+1] = y[n] + h * y[n] * (1.0 - y[n])
    rem = T - t[-1]
    if rem > 1e-12:
        t = np.append(t, T)
        y = np.append(y, y[-1] + rem * y[-1] * (1.0 - y[-1]))
    return t, y
```

```python
def check_qualitative(y, tol_monotone=1e-12, tol_conv=1e-3):
    diffs = np.diff(y)
    monotone_up = np.all(diffs >= -tol_monotone)
    in_open_01 = np.all((y > 0.0) & (y < 1.0))
    conv_to_1  = abs(y[-1] - 1.0) < tol_conv
    return monotone_up, in_open_01, conv_to_1
```

```python
y0 = 0.5
T  = 10.0
h_list = np.round(np.linspace(0.05, 2.00, 196), 4)
```

```python
rows = []
for h in h_list:
    t, y = forward_euler_logistic(y0, h, T)
    mono, inside, conv = check_qualitative(y)
    rows.append({
        "h": h,
        "monotone_up": bool(mono),
        "stays_in_(0,1)": bool(inside),
        "converges_to_1@T": bool(conv),
        "qualitatively_correct": bool(mono and inside and conv),
        "y_T": float(y[-1]),
        "N_steps": len(t)-1
    })

df = pd.DataFrame(rows)
print(df[df["qualitatively_correct"]].head())
print("Max h passing (grid):", df[df["qualitatively_correct"]]["h"].max())
```

```python
df.to_csv("logistic_forward_euler_h_scan.csv", index=False)

plt.figure()
plt.plot(df["h"].values, df["qualitatively_correct"].astype(int).values)
plt.xlabel("h")
plt.ylabel("qualitatively_correct (1=yes, 0=no)")
plt.title("Forward Euler for y' = y(1-y): qualitative correctness vs h")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("logistic_forward_euler_h_scan_indicator.png")
plt.show()
```

In above we found out that the maximum $h$ to make sure the solution qualitatively correct is $h=0.99$ it is same as we write before.
